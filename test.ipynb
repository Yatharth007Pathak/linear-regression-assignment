```python
import numpy as np
```

```python
# SIMPLE LINEAR REGRESSION

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)

n = len(x)

m = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x*x) - (np.sum(x))**2)
c = (np.sum(y) - m*np.sum(x)) / n

print("Slope:", m)
print("Intercept:", c)
import numpy as np
```

```python
# MULTIPLE LINEAR REGRESSION

X = np.array([
[1, 1, 2],
[1, 2, 1],
[1, 3, 4],
[1, 4, 3]
], dtype=float)


Y = np.array([6, 8, 18, 16], dtype=float)


beta = np.linalg.inv(X.T @ X) @ X.T @ Y
print("Coefficients:", beta)
```

```python
# COST FUNCTION (Mean Squared Error)

def cost_function(y, y_pred):
    n = len(y)
    return (1 / (2 * n)) * np.sum((y - y_pred) ** 2)
```

```python
# LINEAR REGRESSION USING GRADIENT DESCENT

# Initialize
m = 0
c = 0
alpha = 0.01
epochs = 1000

for _ in range(epochs):
    y_pred = m * x + c
    dm = (-2/n) * np.sum(x * (y - y_pred))
    dc = (-2/n) * np.sum(y - y_pred)

    m = m - alpha * dm
    c = c - alpha * dc

print("Final m:", m)
print("Final c:", c)
```

```python
# Polynomial Regression (Using Linear Algebra)

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([1, 4, 9, 16, 25], dtype=float)


X = np.column_stack((np.ones(len(x)), x, x**2))
beta = np.linalg.inv(X.T @ X) @ X.T @ y


print("Polynomial Coefficients:", beta)
```
